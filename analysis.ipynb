- **dates_df:** Contains date-related information (e.g., date keys, calendar dates, day, month, year, etc.).

```python
bookings_df['revenue_generated'].max()
```

```python
#using std deviation method
mean = bookings_df['revenue_generated'].mean()
std = bookings_df['revenue_generated'].std()
high = mean + 3 * std
low = mean - 3 * std
#bookings_df=bookings_df[(bookings_df['revenue_generated'] > low) & (bookings_df['revenue_generated']<high)]
bookings_df['revenue_generated']= bookings_df['revenue_generated'].clip(lower=low, upper=high).round(2)
```

```python
bookings_df.revenue_generated.max()
```

```python
bookings_df["revenue_realized"].max()
```

```python
#using std deviation method
mean = bookings_df['revenue_realized'].mean()
std = bookings_df['revenue_realized'].std()
high = mean + 3 * std
low = mean - 3 * std
bookings_df=bookings_df[(bookings_df['revenue_realized'] > low) & (bookings_df['revenue_realized']<high)]
```

### ==> 3. Data Transformation
***

**Create occupancy percentage column**

```python
room_bookings_agg_df.head()
```

```python
room_bookings_agg_df['occ_percent'] = room_bookings_agg_df.apply(
    lambda row: (row['successful_bookings'] / row['capacity']) * 100, axis=1
).round(1)
```

```python
room_bookings_agg_df.head()
```

```python
room_bookings_agg_df.shape
```

```python
property_occupancy_df = property_occupancy_df.rename(columns={'occ%': 'occ_percent'})
```

```python
property_occupancy_df.head()
```

```python
merged_df = pd.merge(
    property_occupancy_df,
    room_bookings_agg_df,
    on=['property_id', 'check_in_date', 'room_category'],
    how='outer',  # use 'inner' if you want only matching rows
    suffixes=('_property', '_room')
)
```

## Average revenue by room_category

```python
# Calculate average real revenue by room_category
avg_revenue_by_category = bookings_df.groupby('room_category')['real_revenue'].mean().reset_index(name='avg_real_revenue')

# Merge with rooms_df to get room_class
avg_revenue_with_class = avg_revenue_by_category.merge(rooms_df, left_on='room_category', right_on='room_id', how='left')

# Drop the redundant room_id column if you want
avg_revenue_with_class = avg_revenue_with_class.drop(columns=['room_id'])

pd.DataFrame(avg_revenue_with_class )
```

## Number of Bookings in all Platform

```python
result = bookings_df.groupby('booking_platform').size().reset_index(name='count').sort_values('count', ascending=False)
result
```

## 2. Average occupancy rate in each of the room categories?

```python
avg_occ = room_bookings_agg_df.groupby('room_category')['occ_percent'].mean().reset_index(name='avg_occ_percent')

# Merge with rooms_df to get room_class
avg_occ_with_class = avg_occ.merge(rooms_df, left_on='room_category', right_on='room_id', how='left')

# Drop the redundant room_id column if you want
avg_occ_with_class = avg_occ_with_class.drop(columns=['room_id'])

avg_occ_with_class
```

```python
merged_dfs= bookings_df.merge(hotels_df ,left_on ='property_id' ,right_on='property_id',how='left')
merged_dfs.head()
```
